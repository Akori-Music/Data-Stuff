```python
#5 Print the row from the dataset based on the most expensive food item in Price in USD.
df[df['Price in USD'] == df['Price in USD'].max()]
```

```python
#6 Print the row from the dataset based on the cheapest food item in Price in USD.
df[df['Price in USD'] == df['Price in USD'].min()]
```

```python
#7 Find the highest prices in USD for all the four food items
df.groupby('Food Item ')['Price in USD'].max()
```

```python
#8 What is the average Price in USD of food in all the five countries from 2018 to 2022?
# PLACE YOUR CODES IN A NEW CELL SO THAT THE TABLE BELOW CAN BE USED FOR COMPARISION.
```

```python
df.groupby(['Year ','Country '])[['Month ', 'Average Price ', 'Price in USD', 'Availability ']].mean()
```

```python
#9 Based on the previous table (#8), create line relational plots to show the increase of average food Price in USD from 2018 to 2020 for each of the five countries (see below).
# sns.FacetGrid() IS NOT REQUIRED!
# PLACE YOUR CODES IN A NEW CELL SO THAT THE GRAPHS BELOW CAN BE USED FOR COMPARISION.
```

```python
#10 What are the median price in USD of Eggs in Japan for the years 2018, 2019, 2020, 2021 and 2022 separately?
df[df['Country '] == 'Japan'].groupby('Year ')['Price in USD'].median()
```

```python
#11 Show the occurences/instances when the Average Prices of Potatoes in Australia during 2020 at above 3.61?
df[(df['Food Item '] == 'Potatoes') & (df['Country '] == 'Australia') & (df['Year '] == 2020) & (df['Average Price '] > 3.61)]
```

```python
#12 How many times were the Bread prices in USD higher or equal to 2.05?
len(df[df['Food Item '] == 'Bread'][df['Price in USD'] >= 2.05])
```
